```python
import time
from matplotlib import pyplot as plt 
#define each varibale with it's domain
#تحديد المتغيرات و المجالات المتاحه لهم
Each_var_with_dom = {
    'CS_140': ['Dr.Ali'],
    'CS_370': ['Dr.Hamed','Dr.Ali'],
    'CS_330': ['Dr.Omar','Dr.Hamed','Dr.Ali' ],
    'CS_391': ['Dr.Omar','Dr.Hamed','Dr.Ali' ],
    'CS_401': ['Dr.Hamed','Dr.Ali']
}
print("Initial State")
print(Each_var_with_dom )

#define the constraints between each variables
#القيود الموجوده بين المتغيرات
constraints = {
    ('CS_140', 'CS_370'): lambda a, b: a != b,
    ('CS_370', 'CS_140'): lambda a, b: a != b,
    ('CS_370', 'CS_330'): lambda b, a: b != a,
    ('CS_330', 'CS_370'): lambda b, c: b != c,
    ('CS_330', 'CS_391'): lambda c, b: c != b,
    ('CS_391', 'CS_330'): lambda c, b: c != b,
    ('CS_330', 'CS_401'): lambda c, b: c != b,
    ('CS_401', 'CS_330'): lambda c, b: c != b,
    ('CS_391', 'CS_370'): lambda c, b: c != b,
    ('CS_370', 'CS_391'): lambda c, b: c != b,
    ('CS_391', 'CS_401'): lambda c, b: c != b,
    ('CS_401', 'CS_391'): lambda c, b: c != b,

}

```

<br><br>
<p style="text-align:lift;font-size:20px;"><font color='00BFFF'>Forword-Checking Algorithms</font></p>
<br>


                                                                                             


<br><br>

```python
#Make all courses are consistent
#جعل جميع المواد مستوفين للقيود
def Forword_Checking(Check_Courses):
    start = time.perf_counter_ns()
    #Chose the first value from each course
    #اختار القيمة الاوله من الماده
    for key in Each_var_with_dom :
        Each_var_with_dom[key] = Each_var_with_dom[key][0]

        #put all arcs in queue
        #وضع جميع المواد في طابور
        queue = Check_Courses[:]

        #Check if the first value of each course are consistent with the courses that have crossbond with it course
        #التاكد من ان القيمه الاوله للمتغيرات مستوفيه للقيود مع المواد اللتي تمتلك قيود معها
        while queue :
            (Course1, Course2)=queue.pop(0)
            Modify(Course1, Course2)
    end = time.perf_counter_ns()
    return end - start
```

```python
#Check between two courses if they are consistent and if it's not then remove the inconsistent course values from second course
# التاكد بين المادتين اذا كانو مستوفين الشروط و اذا لم يستوفون الشروط نقوم بحذف القيمه من المادة الثانية
def Modify(Course1, Course2):

    # Get the domain of Course1 and Course2
    #اخد المجالات من الماده الاوله والثانية
    Course1_domain = Each_var_with_dom[Course1]
    Course2_domain = Each_var_with_dom[Course2]

    #This for loop as a pointer for Course2 values
    #لمرور على قيم الماده الثانية
    for i in Course2_domain :
        #Check if the value of Course1 exist in Course2 then remove it from Course2
        # التاكد اذا قيمة الماده الاوله موجوده في الماده الثانية وحذفها
        if Course1_domain == i :
            Course2_domain.remove(Course1_domain)
```

```python
#define a list that include courses that have crossbond for ForwardChecking
#قائمة تجمع المواد مع المواد اللي معها بالقيود
Check_Courses = [('CS_140','CS_370'), ('CS_370','CS_391'), ('CS_391','CS_401'),
                 ('CS_330','CS_401'), ('CS_370','CS_330'), ('CS_330','CS_391')]
#Solve the problem with ForwardChecking
#حل المشكله باستخدام الفور ورد شيكنق
Forword_Checking_Time = Forword_Checking(Check_Courses)
print(Each_var_with_dom)
```

<br><br>
<p style="text-align:lift;font-size:20px;"><font color='00BFFF'>ARC-3 Algorithms</font></p>
 






<br><br>

```python
#Make all arcs are consistent
#  تحقيق جميع القيود المطلوبه بين المواد
def AC_3(Check_arcs):
    start = time.perf_counter_ns()
    
    #define queue and put all arcs in it
    # وضع جميع المواد في طابور
    queue = Check_arcs[:]

    #while queue not empty
    #الاستمرار حتى انتهاء الطابور
    while queue:
        #take the arcs from the queue and put it in Course1 and Course2
        #اخذ مادتين من الطابور ووضعهم في المادة الاوله والثانية
        (Course1, Course2) = queue.pop(0)

        #Check between Course1 and Course2 if they are consistent and if not then make it consistent
        #التاكد بين المادة الاوله والثانية بانهم مستوفين القيود واذا لم يكونو كذلك نجعلهم مستوفين للقيود
        Is_Consistent = Remove_Inconsistent_Values(Course1, Course2)

        #If Course1 and Course2 are consistent
        #اذا كانو الماده الاوله والثانية مستوفين للشروط
        if Is_Consistent:
            #define neighbors and put all arcs in it
            #سحب جميع الجيران وارجاعهم للطابور
            neighbors = [neighbor for neighbor in Check_arcs if neighbor[1] == Course1]
            queue = queue + neighbors
            
    end = time.perf_counter_ns()
    return end - start
```

```python
#Check between two courses if they are consistent and if it's not then remove the inconsistent course values
# التاكد بين مادتين بان لا يوجد بينهم مشكله في القيود واذا كان فيه مشكله نحذف القيمه اللتي يمنعها القيد
def Remove_Inconsistent_Values(Course1, Course2):

    Is_Consistent = False

    #Get the domain of Course1 and Course2
    # اخذ مجالات الماده الاوله والثانية
    Course1_domain = Each_var_with_dom[Course1]
    Course2_domain = Each_var_with_dom[Course2]

    #Get all arc (Course1, Course2) constraints
    #اخذ جميع قيود الماده الاوله والثانيه
    all_constraints = [
        constraint for constraint in constraints if constraint[0] == Course1 and constraint[1] == Course2]
    #This for loop as a pointer for Course1 values
    # المرور على القيم المتاحه للماده الاوله
    for Course1_value in Course1_domain:
        satisfies = False
        #This for loop as a pointer for Course2 values
        #المرور على القيم المتاحه للمادة الثانية
        for Course2_value in Course2_domain:
            #This for loop will check if Course1 and Course2 are satisfies
            #المرور على جميع القيود بين الماده الاوله والثانيه وتاكد بانهم يستفون الشروط
            for constraint in all_constraints:
                constraint_func = constraints[constraint]
                if constraint_func(Course1_value, Course2_value):
                    satisfies = True
        #If Course1 and Course2 are not satisfies then remove the value that make them not satisfies
        #اذا لم تكن المادة الاوله او الثانيه تستوفي الشروط نحذف القيمه اللتي جعلتها لم تستوفي الشروط
        if not satisfies:
            Course1_domain.remove(Course1_value)
            Is_Consistent = True

    return Is_Consistent
```

```python
#define a list that include all courses that have crossbond for ARC_3
#قائمة تجمع المواد مع المواد اللي معها بالقيود
Check_arcs = [('CS_140','CS_370'), ('CS_370','CS_140'), ('CS_370','CS_330'), ('CS_330','CS_370'), ('CS_330','CS_391'),
              ('CS_391','CS_330'), ('CS_330','CS_401'), ('CS_401','CS_330'), ('CS_391','CS_370'), ('CS_370','CS_391'),
              ('CS_391','CS_401'), ('CS_401','CS_391')]

#Solve the problem with ARC_3
#حل المشكله باستخدام الارك
AC_3_Time = AC_3(Check_arcs)
print(Each_var_with_dom)
```
